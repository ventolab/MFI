```python
%%R -o overlap_IDs

# what is the overlap of current cell IDs and the ones we're trying to subset?

overlap_IDs = Reduce(intersect,  list(rownames(meta_atac), proj$cellNames) )
```

```python
%%R

# subsetting archR object to IDs only in meta_atac

proj <- proj[overlap_IDs, ]
```

```python
%%R

proj
```

```python
overlap_IDs = [overlap_IDs[i] for i in range(len(overlap_IDs))]
```

```python
overlap_IDs[:5]
```

```python
# subsetting metadata

meta_atac = meta_atac.loc[overlap_IDs,:]
```

```python
%%R -i meta_atac

proj <- addCellColData(ArchRProj = proj, data = paste0(meta_atac$annot_multiome_combined),
    cells = overlap_IDs, 
                       name = "annot_multiome_combined"
                      )
```

```python
%%R

# generating bigWig files per pseudobulk
# takes seemingly a lot of RAM (~100Gb ish?) adn also takes VERY long (on the order of hours, not sure it even completes) for larger files/groups

getGroupBW(
  ArchRProj = proj,
  groupBy = "annot_multiome_combined",
  normMethod = "ReadsInTSS",
  tileSize = 100,
  maxCells = 1000,
  ceiling = 4,
  verbose = TRUE,
  threads = getArchRThreads(),
  logFile = createLogFile("getGroupBW")
)
```
